```python
with open ("archivos_json/definitivo.json", "r") as fichero:
    definitivo = json.load(fichero)

lista_flamenco = []
canciones_totales = lista_flamenco

def extraer_por_genero(definitivo, flamenco):
    global lista_flamenco, canciones_totales
    for item in definitivo:
        año = item.get("year")
        if item.get("genre") == flamenco and 2016 <= año <= 2020:
            info = {"Artista": item.get("artist_name"),
                    "Género": item.get("genre"),
                    "Nombre de la cancion": item.get("track_name"),
                    "Año de lanzamiento": item.get("year")}
             
            lista_flamenco.append(info)
    numero_canciones = len(lista_flamenco)
    print(numero_canciones)
    return canciones_totales 
     
extraer_por_genero(definitivo, "flamenco")
```

```python
def popularidad_artista(nombre_artista):
    url = "http://ws.audioscrobbler.com/2.0/"
    
    params = {
        "method": "artist.getinfo",
        "artist": nombre_artista,
        "api_key": API_KEY,
        "format": "json"
    }
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        
        stats = data.get("artist", {}).get("stats", {})
        
        return {
            "listeners": stats.get("listeners"),
            "playcount": stats.get("playcount")
        }
    else:
        print(response.reason, response.status_code)
```
