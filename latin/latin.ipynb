```python
import json
import requests
import pandas as pd
```

```python
with open ("../archivos_json/definitivo.json", "r") as fichero: 
    definitivo = json.load(fichero)           

def extraer_por_genero(canciones, genero):    
    lista_latin = []                      

    for item in canciones:                  
        año = item.get("year")

        if item.get("genre") == genero and 2016 <= año <= 2019:    
            info = (                                       
                item.get("artist_name"),
                item.get("genre"),
                item.get("track_name"),
                año
            )

            lista_latin.append(info)        

    return lista_latin                   

lista_latin = extraer_por_genero(definitivo, "latin")      
print(f"La longitud de la lista es ---> {len(lista_latin)} canciones de Latin")
print(f"{'-' * 100} \n{lista_latin}")
```

```python
artistas_latin = set()                 
for dato in lista_latin:                       
    artistas_latin.add(dato[0])       
print(f"La longitud de los artistas es ---> {len(artistas_latin)} artistas \n{'-' * 100} \nLos artistas son ---> {artistas_latin}")  
```

```python
print(f"La lista del genero Latin ---> {lista_latin}")     
lista_artistas = list(artistas_latin)                     
print("--------------------------------------")
print(f"Los artistas en Latin son: {lista_artistas} con {len(lista_artistas)} artistas")
```

```python
df_latin = pd.DataFrame(lista_latin, columns= ["artist", "genre", "track_name", "year"])
```

```python
API_KEY = "23b77816bcd74ca99bfab67b0e314dd0"

def info_lastfm(artista):              
    url = "https://ws.audioscrobbler.com/2.0/"            
    
    params = {                              
        "method": "artist.getinfo", 
        "artist": artista,
        "api_key": API_KEY,
        "format": "json"
    }
    
    response = requests.get(url, params=params)                  
    if response.status_code != 200:                 
        print(f"Error en la petición")
        return None
    
    data = response.json()                      
    
    artist = data["artist"]                
    
    return (                        
        artista,                 
        artist["bio"]["summary"], 
        artist["stats"]["listeners"],
        artist["stats"]["playcount"],
        [a["name"] for a in artist["similar"]["artist"]]
    )          

resultados = []
errores = []

for artista in lista_artistas:      
    try:
        info = info_lastfm(artista)     
        resultados.append(info)
    except Exception as e:                   
        errores.append({
            "artista": artista,
            "error": str(e)
        })
        continue        
```

```python
print(f"Resultado de la busqueda --> {resultados} \n {"-" * 100} \nLa longitud de los resultados --> {len(resultados)} \n {"-" * 100} \nErrores encontrados --> {errores}")
```

```python
df_resultado = pd.DataFrame(resultados, columns= ["artist", "bio", "listeners", "playcount", "similar_artist"])
```
