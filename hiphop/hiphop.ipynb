```python
import json
import requests
import pandas as pd
```

```python
with open ("../archivos_json/definitivo.json", "r") as fichero:
    definitivo = json.load(fichero)

def extraer_por_genero(canciones, genero):
    lista_hiphop = []

    for item in canciones:
        año = item.get("year")

        if item.get("genre") == genero and 2016 <= año <= 2019:
            info = {
                "Artista": item.get("artist_name"),
                "Género": item.get("genre"),
                "Nombre de la Canción": item.get("track_name"),
                "Año de lanzamiento": año
            }

            lista_hiphop.append(info)

    print(len(lista_hiphop))
    return lista_hiphop

lista_hiphop = extraer_por_genero(definitivo, "hip-hop")
print(lista_hiphop[0:5])
```

```python
artistas_hiphop = set()
for artista in lista_hiphop:
    artistas_hiphop.add(artista["Artista"])
print(f" {len(artistas_hiphop)}")
```

```python
with open ("lista_hiphop.json", "w", encoding="utf-8") as fichero:
    json.dump(lista_hiphop, fichero, ensure_ascii=False, indent=4)
```
