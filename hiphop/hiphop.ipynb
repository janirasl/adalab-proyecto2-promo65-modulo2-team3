```python
with open("../archivos_json/definitivo.json", "r", encoding="utf-8") as fichero:  # Ruta al json
    definitivo = json.load(fichero)  # Abrimos el fichero


def extraer_por_genero(canciones, genero):
    # Creamos la funcion
    lista_hiphop = []  # lista vacia para almacenar los datos extraidos

    for item in canciones:  # iteramos por la lista
        año = item.get("year")

        # Normalizamos el género por si viene como "Hip-Hop", "hip-hop", etc.
        genero_item = (item.get("genre") or "").strip().lower().replace("-", " ")
        genero_objetivo = genero.strip().lower().replace("-", " ")

        if genero_item == genero_objetivo and isinstance(año, int) and 2016 <= año <= 2019:
            info = (
                item.get("artist_name"),
                item.get("genre"),
                item.get("track_name"),
                año
            )
            lista_hiphop.append(info)

    return lista_hiphop


lista_hiphop = extraer_por_genero(definitivo, "hip hop")
print(f"La longitud de la lista es --> {len(lista_hiphop)} canciones de Hip Hop")
print(f"{'-' * 100}\n{lista_hiphop}")
```

```python
API_KEY = "e8a8b4cd33b86332f741d2fd683ccabb"  # <-- pon tu API key aqui

def info_lastfm(artista):
    # Creamos una funcion que si le damos el nombre de un artista devuelve su info de Last.fm
    url = "https://ws.audioscrobbler.com/2.0/"  # endpoint

    params = {
        "method": "artist.getinfo",
        "artist": artista,
        "api_key": API_KEY,
        "format": "json"
    }

    response = requests.get(url, params=params, timeout=20)

    if response.status_code != 200:
        print("Error en la petición")
        return None

    data = response.json()

    # A veces Last.fm devuelve {"error":..., "message":...}
    if "artist" not in data:
        raise KeyError("artist")

    artist = data["artist"]

    return (
        artista,
        artist.get("bio", {}).get("summary"),
        artist.get("stats", {}).get("listeners"),
        artist.get("stats", {}).get("playcount"),
        [a.get("name") for a in artist.get("similar", {}).get("artist", [])]
    )


resultados = []
errores = []

for artista in lista_artistas:  # Bucle para que itere por la lista de artistas
    try:
        info = info_lastfm(artista)  # aplicar la funcion
        if info is not None:
            resultados.append(info)
    except Exception as e:
        errores.append({
            "artista": artista,
            "error": str(e)
        })
        continue
```

```python
print(
    f"Resultado de la busqueda --> {resultados} \n"
    f"{'-' * 100}\n"
    f"La longitud de los resultados --> {len(resultados)} \n"
    f"{'-' * 100}\n"
    f"Errores encontrados --> {errores}"
)
```

```python
df_resultado = pd.DataFrame(resultados, columns=["artist", "bio", "listeners", "playcount", "similar_artist"])
```
