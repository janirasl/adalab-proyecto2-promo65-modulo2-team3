```python
#Importamos las librerias necesarias
import json
import requests
import pandas as pd
import requests
```

```python
with open ("../archivos_json/definitivo.json", "r") as fichero: #Ruta hacia el json del que queremos extraer la info
    definitivo = json.load(fichero)           #Abrimos el fichero

def extraer_por_genero(canciones, genero):     #Creamos la funcion
    lista_rock = []                      #lista vacia para almacenar los datos extraidos

    for item in canciones:                  #iteramos por la lista, accedemos al diccionario y al value que necesitamos
        año = item.get("year")

        if item.get("genre") == genero and 2016 <= año <= 2019:     #si cumple la condicion
            info = (                                       #Añadimos a una tupla los datos que queremos
                item.get("artist_name"),
                item.get("genre"),
                item.get("track_name"),
                año
            )

            lista_rock.append(info)           #Añadimos cada iteracion por el diccionario a la lista

    return lista_rock                    #Nos devuelve la lista completa

lista_rock = extraer_por_genero(definitivo, "rock")      #Usamos la funcion para acceder al genero "rock"
print(f"La longitud de la lista es ---> {len(lista_rock)} canciones de Rock")
print(f"{'-' * 100} \n{lista_rock}")
```

```python
artistas_rock = set()                 #Creamos un set para saber los artistas dentro de la lista sin que se repitan
for dato in lista_rock:                        #Iteramos por la lista para acceder a cada una de las tuplas
    artistas_rock.add(dato[0])         #Accedemos al indice 0 donde se almacena el artista que necesitamos y lo añadimos al set vacio
print(f"La longitud de los artistas es ---> {len(artistas_rock)} artistas \n{'-' * 100} \nLos artistas son ---> {artistas_rock}")  
```

```python
print(f"La lista del genero rock ---> {lista_rock}")     
lista_artistas = list(artistas_rock)                      #Convertimos el set en una lista para poder iterar por los datos
print("--------------------------------------")
print(f"Los artistas en rock son: {lista_artistas} con {len(lista_artistas)} artistas")
```

```python
with open ("lista_rock.json", "w", encoding="utf-8") as fichero:       #Guardamos los datos extraidos del genero en un archivo tipo json
    json.dump(lista_rock, fichero, ensure_ascii=False, indent=4)      #Usamos metodos para que sea mas legible
```

```python
with open("artistas_rock.json", "w", encoding="utf-8") as fichero:
    json.dump(sorted(artistas_rock),                              #Guardamos los datos extraidos de los artistas en un archivo tipo json
        fichero,
        ensure_ascii=False,                                       #Lo hacemos mas legible
        indent=4)
```
