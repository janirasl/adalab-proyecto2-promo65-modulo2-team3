```python
with open ("../archivos_json/definitivo.json", "r") as fichero:
    definitivo = json.load(fichero)
def extraer_por_genero(canciones, genero):
    lista_reggaeton = []
    for item in canciones:
        año = item.get("year")
        if item.get("genre") == genero and 2016 <= año <= 2019:
            info = {
                "Artista": item.get("artist_name"),
                "Género": item.get("genre"),
                "Nombre de la Canción": item.get("track_name"),
                "Año de lanzamiento": año
            }
            lista_reggaeton.append(info)
    print(len(lista_reggaeton))
    return lista_reggaeton
lista_reggaeton = extraer_por_genero(definitivo, "reggaeton")
print(lista_reggaeton)
artistas = set()
for artista in lista_reggaeton:
    artistas.add(artista["Artista"])
print(f" {len(artistas)} {artistas}")
```

```python

API_KEY = API_KEY # <-- pon tu API key aquí

def info_lastfm(artista):              #Creamos una funcion que si le damos el nombre de un artista devuelve su info de Last.fm”
    url = "https://ws.audioscrobbler.com/2.0/"            #le decimos de donde coger la info
    
    params = {                              #rellenamos el formulario con lo que le pedimos
        "method": "artist.getinfo", 
        "artist": artista,
        "api_key": API_KEY,
        "format": "json"
    }
    
    response = requests.get(url, params=params)                   #le hacemos la peticion
    if response.status_code != 200:                  #Verificamos que la instruccion es correcta
        print(f"Error en la petición")
        return None
    
    data = response.json()                      #convertimos la respuesta en un diccionario
    
    artist = data["artist"]                #le decimos la key dentro de data
    
    return (                         #lo que la funcion nos va a devolver
        artist["mbid"],
        artist["name"],                
        artist["bio"]["summary"], 
        artist["stats"]["listeners"],
        artist["stats"]["playcount"],
        artist["similar"]["artist"][0]["name"],
    )               #“De cada artista similar (a), quédate solo con su nombre”

resultados = []
errores = []

for artista in lista_artistas:        #Bucle para que itere por la lista de artistas
    try:
        info = info_lastfm(artista)      #Que intente aplicar la funcion para cada artista de la lista
        resultados.append(info)
    except Exception as e:                    #Si no lo logra que lo guarde en un error
        errores.append({
            "artista": artista,
            "error": str(e)
        })
        continue        #Sigue con el siguiente artista
```

```python
df_resultado = pd.DataFrame(resultados, columns= ["id_artist", "artist", "bio", "listeners", "playcount", "similar_artist"])
```
