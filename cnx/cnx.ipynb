```python
#Importamos las librerias necesarias
import requests
from dotenv import load_dotenv 
import os
import mysql.connector
import pandas as pd

load_dotenv("../hiphop/.env")
SQL_pass = os.getenv("SQL")
```

```python
def subir_csv_a_sql(ruta_csv):

    df = pd.read_csv(ruta_csv)                              # Leer el CSV

    conexion = mysql.connector.connect(
        host="127.0.0.1",                                  # Conectar a MySQL
        user="root",
        password=SQL_pass,
        database="music_db"
    )

    cursor = conexion.cursor() 

    errores = []

    for index, fila in df.iterrows():                 # Recorrer cada fila del dataframe     
        try:  
            #query para insertar los datos
            query = """
            INSERT INTO artists                                     
            (id_artist, artist_name, artist_bio, artist_listeners, artist_playcount, similar_artist)
            VALUES (%s, %s, %s, %s, %s, %s)
            """

            valores = (
                fila["id_artist"],
                fila["artist"],
                fila["bio"],
                fila["listeners"],
                fila["playcount"],
                fila["similar_artist"]
            )

            cursor.execute(query, valores)

        except Exception as e:                #Guardar el error y la fila
            errores.append((index, str(e)))
            print(f"Fila {index} fallo: {e}")
            continue                              #Sigue

    conexion.commit()                                    # Guardar cambios

    cursor.close()                                      # Cerrar conexión
    conexion.close()

    print("Datos subidos correctamente")
```

```python
def subir_csv_a_sql_tracks(ruta_csv):

    df = pd.read_csv(ruta_csv)                              # Leer el CSV

    mapeo_generos = {"rock": 1,"indie": 2,"latin": 3,"reggaeton": 4,"hip-hop": 5}

    conexion = mysql.connector.connect(
        host="127.0.0.1",                                  # Conectar a MySQL
        user="root",
        password=SQL_pass,
        database="music_db"
    )

    cursor = conexion.cursor() 

    errores = []

    for index, fila in df.iterrows():                 # Recorrer cada fila del dataframe     
        try: 
            nombre_genero_csv = fila["genre"] 
            id_genero_numerico = mapeo_generos.get(nombre_genero_csv, None)             
            if id_genero_numerico is None:                 
                print(f"Fila {index}: El género '{nombre_genero_csv}' no está en el mapeo. Saltando...") 
                continue

            #query para insertar los datos
            query = """
            INSERT INTO tracks                                     
            (id_artist, id_genre, track_name, year)
            VALUES (%s, %s, %s, %s)
            """

            valores = (
                fila["artist"],
                id_genero_numerico,
                fila["track_name"],
                fila["year"],
            )

            cursor.execute(query, valores)

        except Exception as e:                #Guardar el error y la fila
            errores.append((index, str(e)))
            print(f"Fila {index} fallo: {e}")
            continue                              #Sigue

    conexion.commit()                                    # Guardar cambios

    cursor.close()                                      # Cerrar conexión
    conexion.close()

    print("Datos subidos correctamente")
```

```python
def subir_csv_a_sql_tracks(ruta_csv):
    # 1. Leer el CSV
    df = pd.read_csv(ruta_csv)
    # Convertir NaNs a None para que SQL no explote
    df = df.where(pd.notnull(df), None)

    # El mapeo debe coincidir EXACTAMENTE con lo que hay en tu tabla 'genre'
    mapeo_generos = {"rock": 1, "indie": 2, "latin": 3, "reggaeton": 4, "hip-hop": 5}

    try:
        conexion = mysql.connector.connect(
            host="127.0.0.1",
            user="root",
            password=SQL_pass, # Asegúrate de que esta variable sea accesible
            database="music_db"
        )
        cursor = conexion.cursor()
        
        # Opcional: Obtener los mbid válidos de la DB para verificar antes de insertar
        cursor.execute("SELECT id_artist FROM artists")
        artistas_validos = {row[0] for row in cursor.fetchall()}

        errores = []

        for index, fila in df.iterrows():
            try:
                # 1. Manejo del género (Normalizamos a minúsculas por si acaso)
                nombre_genero_csv = str(fila["genre"]).lower() if fila["genre"] else ""
                id_genero_numerico = mapeo_generos.get(nombre_genero_csv)
                
                # 2. Obtener el ID del artista (mbid)
                id_artista_csv = fila["artist"] # Este debe ser el MBID que subiste antes

                # Validaciones previas
                if id_genero_numerico is None:
                    print(f"Fila {index}: Género '{nombre_genero_csv}' no reconocido.")
                    continue
                
                if id_artista_csv not in artistas_validos:
                    print(f"Fila {index}: El artista '{id_artista_csv}' no existe en la tabla 'artists'.")
                    continue

                # 3. Query de inserción (Fíjate en los nombres de columnas de tu imagen)
                query = """
                INSERT INTO tracks 
                (id_artist, id_genre, track_name, year)
                VALUES (%s, %s, %s, %s)
                """
                
                valores = (
                    id_artista_csv,     # Debe ser VARCHAR(50) según tu diagrama
                    id_genero_numerico, # INT
                    fila["track_name"], # VARCHAR
                    fila["year"]        # YEAR
                )

                cursor.execute(query, valores)

            except Exception as e:
                errores.append((index, str(e)))
                print(f"Error en fila {index}: {e}")

        conexion.commit()
        print(f"Proceso terminado. Se intentaron subir {len(df)} canciones.")

    except mysql.connector.Error as err:
        print(f"Error de conexión: {err}")
    finally:
        if 'conexion' in locals() and conexion.is_connected():
            cursor.close()
            conexion.close()

# Llamada a la función
# subir_csv_a_sql_tracks("mis_canciones.csv")
```

```python
def insertar_cancion_con_genero(conexion, titulo, nombre_genero): 

    cursor = conexion.cursor()     # 1. Buscamos el ID del género en la tabla de géneros    

    query_genero = "SELECT id_genero FROM generos WHERE nombre_genero = %s"   

    cursor.execute(query_genero, (nombre_genero,))    

    resultado = cursor.fetchone()     

    if resultado:         
        id_genero_encontrado = resultado[0] # 2. Insertamos la canción usando ese ID 

        query_cancion = """ INSERT INTO canciones (titulo, id_genero) VALUES (%s, %s) """ 

        cursor.execute(query_cancion, (titulo, id_genero_encontrado)) 

        conexion.commit() 
        
    else: print(f"El género '{nombre_genero}' no existe en la base de datos.") 
    cursor.close()
 
```
