```python
import json
import requests
import pandas as pd
```

```python
with open ("archivos_json/definitivo.json", "r") as fichero:
    definitivo = json.load(fichero)

def extraer_por_genero(canciones, genero):
    lista_latin = []

    for item in canciones:
        año = item.get("year")

        if item.get("genre") == genero and 2016 <= año <= 2019:
            info = {
                "Artista": item.get("artist_name"),
                "Género": item.get("genre"),
                "Nombre de la Canción": item.get("track_name"),
                "Año de lanzamiento": año
            }

            lista_latin.append(info)

    print(len(lista_latin))
    return lista_latin

lista_latin = extraer_por_genero(definitivo, "latin")
print(lista_latin)
```

```python
artistas = set()
for artista in lista_latin:
    artistas.add(artista["Artista"])
    resultado_artistas = "\n".join(artistas)
print(len(resultado_artistas))
print(resultado_artistas)
```

```python
def popularidad_artista(nombre_artista):
    url = "http://ws.audioscrobbler.com/2.0/"
    
    params = {
        "method": "artist.getinfo",
        "artist": nombre_artista,
        "api_key": API_KEY,
        "format": "json"
    }
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        
        stats = data.get("artist", {}).get("stats", {})
        
        return {
            "listeners": stats.get("listeners"),
            "playcount": stats.get("playcount")
        }
    else:
        print(response.reason, response.status_code)
```
