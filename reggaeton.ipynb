```python
import json
```

```python
with open ("archivos_json/definitivo.json", "r") as fichero:
    definitivo = json.load(fichero)
def extraer_por_genero(canciones, genero):
    lista_reggaeton = []
    for item in canciones:
        año = item.get("year")
        if item.get("genre") == genero and 2016 <= año <= 2019:
            info = {
                "Artista": item.get("artist_name"),
                "Género": item.get("genre"),
                "Nombre de la Canción": item.get("track_name"),
                "Año de lanzamiento": año
            }
            lista_reggaeton.append(info)
    print(len(lista_reggaeton))
    return lista_reggaeton
lista_reggaeton = extraer_por_genero(definitivo, "reggaeton")
print(lista_reggaeton)
artistas = set()
for artista in lista_reggaeton:
    artistas.add(artista["Artista"])
print(f" {len(artistas)} {artistas}")
```
