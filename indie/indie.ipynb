```python
#importamos las librerías necesarias 
import json
import requests
import pandas as pd
from dotenv import load_dotenv
import os
from bs4 import BeautifulSoup
```

```python
with open ("../archivos_json/definitivo.json", "r") as fichero: #Ruta hacia el json del que queremos extraer la info
    definitivo = json.load(fichero)           #Abrimos el fichero

def extraer_por_genero(canciones, genero):     #Creamos la funcion para que extraiga el genero que queremos 
    lista_indie = []                      #lista vacia para almacenar los datos extraidos

    for item in canciones:                  #iteramos por la lista, accedemos al diccionario y al value que necesitamos
        año = item.get("year")

        if item.get("genre") == genero and 2016 <= año <= 2019:     #si cumple la condicion solo entran canciones del género y año que queremos 
            info = (                                       #Añadimos a una tupla los datos que queremos que se guarden en el diccionario 
                item.get("artist_name"),
                item.get("genre"),
                item.get("track_name"),
                año
            )

            lista_indie.append(info)           #Añadimos cada iteracion por el diccionario a la lista

    return lista_indie                    #Nos devuelve la lista completa

lista_indie = extraer_por_genero(definitivo, "indie")      #Usamos la funcion para acceder al genero "indie"
print(f"La longitud de la lista es ---> {len(lista_indie)} canciones de Indie")
print(f"{'-' * 100} \n{lista_indie}")
```

```python
artistas_indie = set()                 #Creamos un set para saber los artistas dentro de la lista sin que se repitan
for dato in lista_indie:                        #Iteramos por la lista para acceder a cada una de las tuplas
    artistas_indie.add(dato[0])         #Accedemos al indice 0 donde se almacena el artista que necesitamos y lo añadimos al set vacio
print(f"La longitud de los artistas es ---> {len(artistas_indie)} artistas \n{'-' * 100} \nLos artistas son ---> {artistas_indie}")  
```

```python
print(f"La lista del género indie ---> {lista_indie}")     
lista_artistas = list(artistas_indie)                      #Convertimos el set en una lista para poder iterar por los datos
print("--------------------------------------")
print(f"Los artistas en indie son: {lista_artistas} con {len(lista_artistas)} artistas")
```

```python
df_indie = pd.DataFrame(lista_indie, columns= ["artist", "genre", "track_name", "year"]) #convertimos la lista en un dataframe 
```

```python
Api_key = os.getenv("API_KEY")
```

```python
def info_lastfm(artista):              #Creamos una funcion que si le damos el nombre de un artista devuelve su info de Last.fm”
    url = "https://ws.audioscrobbler.com/2.0/"            #le decimos de donde coger la info
    
    params = {                              #rellenamos el formulario con lo que le pedimos
        "method": "artist.getinfo", 
        "artist": artista,
        "api_key": Api_key,
        "format": "json"
    }
    
    response = requests.get(url, params=params)                   #le hacemos la peticion
    if response.status_code != 200:                  #Verificamos que la instruccion es correcta
        print(f"Error en la petición")
        return None
    
    data = response.json()                      #convertimos la respuesta en un diccionario
    
    artist = data.get("artist")
    
    return (  
        artist["mbid"],                      #lo que la funcion nos va a devolver
        artista,                 
        artist["bio"]["summary"], 
        artist["stats"]["listeners"],
        artist["stats"]["playcount"],
        artist["similar"]["artist"][0]["name"]
    )               #“De cada artista similar (a), quédate solo con su nombre”

resultados = []
errores = []

for artista in lista_artistas:        #Bucle para que itere por la lista de artistas
    try:
        info = info_lastfm(artista)      #Que intente aplicar la funcion para cada artista de la lista
        resultados.append(info)
    except Exception as e:                    #Si no lo logra que lo guarde en un error
        errores.append({
            "artista": artista,
            "error": str(e)
        })
        continue        #Sigue con el siguiente artista
```

```python
print(f"Resultado de la busqueda --> {resultados} \n {"-" * 100} \nLa longitud de los resultados --> {len(resultados)} \n {"-" * 100} \nErrores encontrados --> {errores}")
```

```python
df_resultado = pd.DataFrame(resultados, columns= ["artist", "id_artist", "bio", "listeners", "playcount", "similar_artist"])
```

```python
def limpiar_html(texto):
    if isinstance(texto, str) and texto.strip():
        soup = BeautifulSoup(texto, "html.parser")
        return soup.get_text()
    return texto 

df_resultado["bio"] = df_resultado["bio"].apply(limpiar_html)
```
