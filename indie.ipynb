```python
import json
```

```python
with open ("archivos_json/definitivo.json", "r") as fichero:
        definitivo = json.load(fichero)

def extraer_por_genero(definitivo, indie):
    lista_indie = []
    for item in definitivo:
        año = item.get("year")
        if item.get("genre") == indie and 2016 <= año <= 2020:
            info = {"Artista": item.get("artist_name"),
                    "Género": item.get("genre"),
                    "Nombre de la cancion": item.get("track_name"),
                    "Año de lanzamiento": item.get("year")}
             
            lista_indie.append(info)
    canciones_totales = lista_indie
    numero_canciones = len(lista_indie)
    print(numero_canciones)
    return canciones_totales 
     
extraer_por_genero(definitivo, "indie")
```
