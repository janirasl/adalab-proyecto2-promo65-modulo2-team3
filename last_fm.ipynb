```python
API_KEY = "23b77816bcd74ca99bfab67b0e314dd0"

def llamar_artista(nombre_artista):

    url = "http://ws.audioscrobbler.com/2.0/"

    params = {
        "method": "artist.getinfo",
        "artist": nombre_artista,
        "api_key": API_KEY,
        "format": "json"
    }
    response = requests.get(url, params = params)
    if response.status_code == 200:
        bio = response.json()
        return bio["artist"]["bio"]
    else:
        print(response.reason, "-->", response.status_code)
```

```python
def llamada_api(url):
    response = requests.get(url, params = params)
    if response.status_code == 200:
        return response.json()
    else:
        print(response.reason, "-->", response.status_code)
```

```python
API_KEY = "23b77816bcd74ca99bfab67b0e314dd0"

def artistas_similares(nombre_artista):

    url = "http://ws.audioscrobbler.com/2.0/"

    params = {
        "method": "artist.getsimilar",
        "artist": nombre_artista,
        "api_key": API_KEY,
        "format": "json"
    }
    response = requests.get(url, params = params)
    if response.status_code == 200:
        artistas = response.json() 
        lista_artistas =[]
        for artista in artistas["similarartists"]["artist"][0:20]:
            lista_artistas.append(artista["name"])
        return lista_artistas.split()
    else:
        print(response.reason, "-->", response.status_code)
```
